```python
import flopy
import matplotlib.pyplot as plt
import numpy as np
```

```python
rch = np.loadtxt('../input-files/data-files/flow-recharge-mult.dat')
```

```python
rchs = np.unique(rch)
rchs
```

```python
fig, axs = plt.subplots(ncols=5, figsize=(10,5))

for i, ax in enumerate(axs):
    rch_new = np.where(rch == rchs[i], rch, np.nan)

    ax.imshow(rch_new)
    ax.set_title(f'RCH Multiplier\n{rchs[i]}')
```

```python
rch_new = np.where(rch == 0.1, rch, 0)
```
